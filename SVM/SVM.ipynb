Support Vector Machine (SVM)

```python
import pandas as pd
from sklearn.datasets import load_digits
digits = load_digits()
```

```python
digits.feature_names
```

```python
digits.target_names
```

```python
digits.data
```

```python
df = pd.DataFrame(digits.data,columns=digits.feature_names)
df.head()
```

```python
df['target'] = digits.target
df.head()
```

```python
df[df.target==0].head()
```

```python
df[df.target==5].head()
```

```python
df['digit_number'] =df.target.apply(lambda x: digits.target_names[x])
df.head()
```

```python
df[45:55]
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df0 = df[df['target'] == 0]  # Filter for digit 0
df1 = df[df['target'] == 1]  # Filter for digit 1

# Plotting pixel_0 vs pixel_1 for two classes
plt.xlabel('Pixel 0 Intensity')
plt.ylabel('Pixel 1 Intensity')
plt.scatter(df0['pixel_0_0'], df0['pixel_0_1'], color="green", marker='+', label='Digit 0')
plt.scatter(df1['pixel_0_0'], df1['pixel_0_1'], color="blue", marker='.', label='Digit 1')
plt.legend()
plt.title('Scatter Plot of Digit 0 and Digit 1')
plt.show()
```

Train Using Support Vector Machine (SVM)

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop(['target','digit_number'], axis='columns')
y = df.target
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
from sklearn.svm import SVC
model = SVC()
```

```python
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

Kernel


```python
model_linear_kernal = SVC(kernel='linear')
model_linear_kernal.fit(X_train, y_train)
```

```python
model_linear_kernal.score(X_test, y_test)
```

```python
model_linear_kernal = SVC(kernel='rbf')
model_linear_kernal.fit(X_train, y_train)
```

```python
model_linear_kernal.score(X_test, y_test)
```

Regularization and Gamma Parameters

```python
model_C = SVC(C=1)
model_C.fit(X_train, y_train)
model_C.score(X_test, y_test)
```

```python
model_C = SVC(C=10)
model_C.fit(X_train, y_train)
model_C.score(X_test, y_test)
```

```python
model_g = SVC(gamma=10)
model_g.fit(X_train, y_train)
model_g.score(X_test, y_test)
```
