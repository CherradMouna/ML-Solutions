```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('canada_per_capita_income.csv')
df
```

```python
%matplotlib inline
plt.xlabel('year')
plt.ylabel('capital_income_(US$)')
plt.scatter(df.iloc[:, 0], df.iloc[:, 1],color='red',marker='+')
```

```python
new_df = df.drop('year',axis='columns')
year = df.year
```

```python
# Create linear regression object
reg = linear_model.LinearRegression()
reg.fit(new_df,year)
```

```python
reg.predict([[2020]])
```

```python
reg.coef_
```

```python
reg.intercept_
```
